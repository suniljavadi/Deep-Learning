```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

def relu(x):
    return np.maximum(0, x)
```

```python
# Sample data
x1 = 3
x2 = 2
y_actual = 17
```

```python
# Hyperparameters
learning_rate = 0.01
epochs = 20
```

```python
# Initialize weights and bias
np.random.seed(0)
w1 = np.random.uniform(1, 10)
w2 = np.random.uniform(1, 10)
bias = np.random.uniform(1, 10)
```

```python
# Lists to store error and predicted values for visualization
errors = []
ypreds = []
```

```python
# Training loop
for _ in range(epochs):
    # Forward pass
    y_pred = x1 * w1 + x2 * w2
    
    # Calculate error
    error = (y_actual - y_pred) ** 2
    errors.append(error)
    ypreds.append(y_pred)
    
    # Backward pass (update weights)
    w1 += learning_rate * 2 * (y_actual - y_pred) * x1
    w2 += learning_rate * 2 * (y_actual - y_pred) * x2
```

```python
# Plot the progression of the error
plt.figure()
plt.plot(range(epochs), errors)
plt.xlabel('Epoch')
plt.ylabel('Error')
plt.title('Error Progression')
plt.show()
```

```python
# Plot the progression of y_pred
plt.figure()
plt.plot(range(epochs), ypreds)
plt.axhline(y=y_actual, color='r', linestyle='--')
plt.xlabel('Epoch')
plt.ylabel('ypred')
plt.title('ypred Progression')
plt.show()
```
